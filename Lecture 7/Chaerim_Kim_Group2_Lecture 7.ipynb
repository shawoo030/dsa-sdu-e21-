Please implement the merge-sort algorithm in a nonrecursive way.

```python
def merge(S1, S2, S):
    """Merge two sorted Python lists S1 and S2 into properly sized list S."""
    i = j = 0
    while i + j < len(S):
        if j == len(S2) or (i < len(S1) and S1[i] < S2[j]):
            S[i+j] = S1[i] # copy ith element of S1 as next item of S
            i += 1
        else:
            S[i+j] = S2[j] # copy jth element of S2 as next item of S
            j += 1
```

```python
#### this is my code ####


def nonRecursive(S):
    '''first, we need to make sorted lists which elements up to 2 .'''

    temS = S
    s = len(S)
    D = {}
    for k in range(s//2):
        temS = S[0:2]
        S1 = S[0]
        S2 = S[1]
        merge([S1],[S2], temS)
    
        S.remove(S1)
        S.remove(S2)
        D[k] = temS

    # put the last element 
    if S != None:
        D[s//2] = S
    
    
    """ second, we need to merge all the sorted lists """

    while len(D) > 1 :
        d = len(D)
    
        # make new merged list
        D[d] = D[0] + D[1]
        merge(D[0], D[1], D[d])
    
        # delete old merged list
        del D[0]
        del D[1]
    
        # if it's all merged
        if len(D) <=1:
            break
    
        # update all keys
        for key in list(D.keys()):
            D[key-2] = D.pop(key)
    
    return(D[2])
        
```
